```python
import os 
import pickle as pkl
import pandas as pd

XCRISP_KLD_PREDICTIONS_F = f"/Users/colm/repos/output/local/model_predictions/OurModel/model_1_v4_RS_1_{dataset}.pkl"
d = pkl.load(open(XCRISP_KLD_PREDICTIONS_F, 'rb'))
```

```python
predicted_freq_df = []
for t in test_sequences:
    if t not in d:
        continue
    ins_arr = pd.Series(d[t]["predicted"], index=d[t]["indels"], name=t)[-21:]
    predicted_freq_df.append(ins_arr)

predicted_freq_df = pd.concat(predicted_freq_df, axis=1).T
```

```python
actual_freq_df = []
for t in test_sequences:
    if t not in d:
        continue
    ins_arr = pd.Series(d[t]["actual"], index=d[t]["indels"], name=t)[-21:]
    actual_freq_df.append(ins_arr)

actual_freq_df = pd.concat(actual_freq_df, axis=1).T
```

```python
(actual_freq_df[["1+A", "1+T", "1+C", "1+G"]] == 0).any(axis=1).all()
```

```python
import numpy as np

common = np.intersect1d(sample_d.index, predicted_freq_df.index)
sample_d = sample_d.loc[common, :]
```

```python
def print_analysis(sample_d, predicted_freq_df, actual_freq_df, nucleotide, position, is_present, ins_seq=None):
    index = nucleotide + str(position)
    samples = sample_d.loc[sample_d[index]==is_present, :].index
    label = "positive" if is_present else "negative"
    len_nucs = len(nucleotide)

    if ins_seq is None:
        ins_seq = (f"{len_nucs}+" + nucleotide)

    overall_mean = actual_freq_df[ins_seq].mean()
    predicted = predicted_freq_df.loc[samples][ins_seq].mean()
    actual = actual_freq_df.loc[samples][ins_seq].mean()
    print(f"For {nucleotide} at position {position}: across {samples.shape[0]} {label} samples, the predicted frequency of {ins_seq} is {predicted}, and the actual frequency is {actual}. Overall mean {overall_mean}")
    return samples


samples = print_analysis(sample_d, predicted_freq_df, actual_freq_df, "A", "17", True)
print_analysis(sample_d, predicted_freq_df, actual_freq_df, "A", "17", False)
print("\n")

print_analysis(sample_d, predicted_freq_df, actual_freq_df, "T", "17", True)
print_analysis(sample_d, predicted_freq_df, actual_freq_df, "T", "17", False)
print("\n")


print_analysis(sample_d, predicted_freq_df, actual_freq_df, "C", "17", True)
print_analysis(sample_d, predicted_freq_df, actual_freq_df, "C", "17", False)
print("\n")


print_analysis(sample_d, predicted_freq_df, actual_freq_df, "G", "17", True)
print_analysis(sample_d, predicted_freq_df, actual_freq_df, "G", "17", False)
print("\n")
```
